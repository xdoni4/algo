```python
k = 26
class Vertex:
    def __init__(self):
        self.to = [0] * k
        self.terminal = False
        
root = Vertex()

def add_string(s):
    v = root
    for ch in s:
        ind = ord(ch) - ord('a')
        if not v.to[ind]:
            v.to[ind] = Vertex()
        v = v.to[ind]
    v.terminal = True

def is_added(s):
    v = root
    for ch in s:
        ind = ord(ch) - ord('a')
        if not v.to[ind]:
            return False
        else:
            v = v.to[ind]
    if v.terminal:
        return True
    else:
        return False

def dfs(v, res, cur):
    for i in range(k):
        if v.to[i]:
            dfs(v.to[i], res, cur + chr(i + ord('a')))
    if v.terminal:
        res.append(cur)
    
    
def complete(s):
    res = []
    res.append(s)
    v = root
    for ch in s:
        ind = ord(ch) - ord('a')
        if not v.to[ind]:
            return res
        else:
            v = v.to[ind]
    dfs(v, res, s)
    return res
    
add_string('cat')
add_string('catharsis')
add_string('caterpillar')
add_string('conic')
add_string('contact')
add_string('cough')
add_string('counter')
add_string('dress')
add_string('drought')
add_string('dolphin')
add_string('dream')
add_string('dreadful')

print(is_added('dreadful'))
print(is_added('dredd'))
print(is_added('cat'))
print(is_added('conic'))
print(is_added('conics'))
print(is_added('dart'))
print(is_added('dreaming'))
print(is_added('dream'))
print('\n')
print(complete('dr'))
print(complete('d'))
print(complete('c'))
print(complete('ca'))
print(complete('co'))
print(complete('cou'))
print(complete('coun'))
print(complete('cc'))
```
